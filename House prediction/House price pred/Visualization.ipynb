```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(1,10,101)
plt.plot(x,np.sin(x),"-")
plt.plot(x,np.cos(x),"--")
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
fig,ax=plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
ax=plt.axes()
ax.plot(x,np.cos(x),'-')
ax.set(xlim=(0,10),ylim=[-2,2],
        xlabel='x',ylabel='cos(x)',
        title='a simple cos plot')
```

```python
#3_Dimensional plots:
x=np.linspace(1,10,15)
y=np.linspace(11,20,15)
a,b=np.meshgrid(x,y)
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
z=f(a,b)
plt.contour(a,b,z)
plt.colorbar()
```
