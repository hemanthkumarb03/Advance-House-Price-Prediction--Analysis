```python
nan_features=[feature for feature in data.columns if data[feature].isnull().sum()>1]
for feature in nan_features:
    print(feature,np.round(data[feature].isnull().sum()/len(data),4),'% is missing percentage')
```

```python
len(nan_features)
```

# Trying to find relation between nan values and dependent variable

```python
for feature in nan_features:
    dataset=data.copy()
    dataset[feature]=np.where(dataset[feature].isnull(),1,0)
    
    dataset.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

# Numerical features

```python
data=data.drop(['Id'],axis=1)
```

```python
numerical_features=[feature for feature in data.columns if data[feature].dtype != 'O']
```

```python
data[numerical_features].head()
```

# Year Features

```python
year_features=[feature for feature in data.columns if 'Yr' in feature or 'Year' in feature]
data[year_features]
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
```

```python
year_features
```

```python
# descrete feature:
descrete_features=[feature for feature in numerical_features if feature not in year_features and len(dataset[feature].unique())<25]
```

```python
descrete_features
```

# Relationship between descrete features and sale price

```python
for feature in descrete_features:
    dataset=data.copy()
    
    dataset.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.xlabel(feature)
    plt.ylabel('saleprice')
    plt.show()
```

```python
continous_features=[feature for feature in numerical_features if feature not in descrete_features + year_features]
```

```python
continous_features
```

```python
for feature in continous_features:
    dataset=data.copy()
    dataset[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.title(feature)
    plt.show()
```

```python
for feature in continous_features:
    dataset=data.copy()
    if 0 in dataset[feature].unique():
        pass
    else:
        dataset[feature]=np.log(dataset[feature])
        dataset['SalePrice']=np.log(dataset['SalePrice'])
        plt.scatter(dataset[feature],dataset['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("saleprice")
        plt.title(feature)
        plt.show()
 
```

# Outliers


```python
for feature in continous_features:
    dataset=data.copy()
    if 0 in dataset[feature].unique() :
        pass
    else:
        dataset[feature]=np.log(dataset[feature])
        dataset.boxplot(feature)
        plt.title(feature)
        plt.show()
```

```python
cat_features=[features for features in data.columns if data[features].dtypes == 'O']
```

```python
for feature in cat_features:
    print('total categories in ',feature,'is',len(data[feature].unique()))
```
